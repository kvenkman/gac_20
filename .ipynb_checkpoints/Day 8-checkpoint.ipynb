# Part One

```python
input_file, test_input_file = 'day8_input', 'day8_test_input'
```

```python
test = False 

file = test_input_file if(test) else input_file

with open(file, 'r') as f:
    lines = [l.strip() for l in f.readlines()]
```

```python
from collections import defaultdict
```

```python
programDict = defaultdict(int)
accumulator = 0
programIndex = 0

while True:
    codeIncrement = 0
    command = lines[programIndex].split()
    code, val = command[0], int(command[1])
    if not programDict[programIndex]:
        programDict[programIndex] += 1
    else:
        print(accumulator)
        break
    
    if (code == 'acc'):
        accumulator += val
        codeIncrement = 1
    if (code == 'jmp'):
        codeIncrement = val
    if(code == 'nop'):
        codeIncrement = 1
        
    programIndex += codeIncrement
    
    
    
```

```python
programDict = defaultdict(int)
accumulator = 0
programIndex = 0
nopList, jmpList, accList = [], [], []

for i, line in enumerate(lines):
    code, val = command[0], int(command[1])
    if (code == 'jmp'):
        jmpList.append(i)
    if(code == 'nop'):
        nopList.append(i)
    

for idx in nopList:
    programCopy = lines.copy()
    command = programCopy[idx].split()
    _, val = command[0], int(command[1])
    programCopy[idx] = 'jmp ' + str(val)

    while programIndex < len(programCopy):
        codeIncrement = 0
        command = lines[programIndex].split()
        code, val = command[0], int(command[1])
        if not programDict[programIndex]:
            programDict[programIndex] += 1
        else:
            break    
    
        if (code == 'acc'):
            accumulator += val
            codeIncrement = 1
        if (code == 'jmp'):
            codeIncrement = val
        if(code == 'nop'):
            codeIncrement = 1
        
        programIndex += codeIncrement    

    if(programIndex >= len(programCopy)):
        finalAccumulator = accumulator
        break

if programIndex < len(programCopy):
    for idx in jmpList:
        programCopy = lines.copy()
        command = programCopy[idx].split()
        _, val = command[0], int(command[1])
        programCopy[idx] = 'jmp ' + str(val)

        while programIndex < len(programCopy):
            codeIncrement = 0
            command = lines[programIndex].split()
            code, val = command[0], int(command[1])
            if not programDict[programIndex]:
                programDict[programIndex] += 1
            else:
                break    

            if (code == 'acc'):
                accumulator += val
                codeIncrement = 1
            if (code == 'jmp'):
                codeIncrement = val
            if(code == 'nop'):
                codeIncrement = 1

            programIndex += codeIncrement    

        if(programIndex >= len(programCopy)):
            finalAccumulator = accumulator
            break        
        
        


print(accumulator)
```
