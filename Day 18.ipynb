# Part One

```python
input_file, test_input_file = 'day18_input', 'day18_test_input'
```

```python
test = False 

file = test_input_file if(test) else input_file

with open(file, 'r') as f:
    lines = [l.strip() for l in f.readlines() if l != "\n"]
```

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def mult(a, b):
    return a*b

def div(a, b):
    return a/b

operatorDict = {"+":add, "-":subtract, "*":mult, "/":div}
```

```python
def evaluateExpression(expression):
    # print(expression)
    idx1 = expression.find("(")
    if(idx1 != -1):
        count = 0
        for i, c in enumerate(expression[idx1:]):
            if c == "(":
                count += 1
            if c == ")":
                count -= 1
            
            if count == 0:
                break
        idx2 = idx1 + i
        return evaluateExpression(expression[:idx1] + str(evaluateExpression(expression[idx1+1:idx2])) + expression[idx2+1:])
    else:
        while(len(expression.split()) != 1):
            expression = expression.split(" ")
            a, o, b = expression[0:3]
            #print(a, o, b, expression[3:])
            f = operatorDict[o]
            expression = str(f(int(a), int(b))) + " " + " ".join(expression[3:])
        
        
        return str(expression.strip())
```

```python
outputSum = 0
for problem in lines:
    while True:
        problem = evaluateExpression(problem).strip()
        if(len(problem.split(" ")) == 1):
            break
        
        
    outputSum += int(problem)
```

```python
print(outputSum)
```
